# 使用 Sieve of Eratosthenes 算法打印质数

```python
def prime_number(num):
    a = np.arange(1,num)
    a_max = int(np.sqrt(len(a)))
    is_primer = np.ones(len(a),dtype=bool)
    is_primer[0] = False
    
    for i in range(a_max):
        if i in a[is_primer]:
            is_primer[(i**2-1)::i] = False
            
    print(a[is_primer])
```

```python
prime_number(100)
```

# 使用 Monte Carlo Method 计算圆周率（蒙特卡洛方法）

```python
def calc_pi(n_dots):
    x = np.random.random(n_dots)
    y = np.random.random(n_dots)
    distance = np.sqrt(x**2 + y**2)
    in_cycle = distance[distance < 1]
    
    print(4 * float(len(in_cycle)/n_dots))
```

```python
calc_pi(10000000)
```

```python
def f(x,y):
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n=256
x=np.linspace(-3,3,n)
y=np.linspace(-3,3,n)
X,Y=np.meshgrid(x,y)

plt.contourf(X,Y,f(X,Y),8,alpha=.76,cmap=plt.cm.hot)
C = plt.contour(X,Y,f(X,Y),8,colors='black',linewidth=.5)
plt.clabel(C,inline=1,fontsize=10)
```

```python
plt.imshow(f(X,Y),cmap='hot',origin='low')
plt.colorbar(shrink=.83)
ax=plt.gca()
ax.grid(which='major',axis='x')
ax.grid(which='major',axis='y')
```
